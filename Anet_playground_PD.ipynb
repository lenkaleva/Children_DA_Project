```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "vscode"
px.defaults.template = "plotly"
```

```python
df = pd.read_csv("data_da.csv")
```

```python
# Nejdriv zafiltrovat sloupce - az pak dolu
df_filter = df.groupby(["YEAR", "SEX"], as_index=False)["OVERWEIGHT"].mean()
df_filter
```

```python
# Zadani df, columns (pripadne uz zafiltrovane) na osy, nazev grafu

df_filter['SEX'] = df_filter['SEX'].map({2: 'Girls', 1: 'Boys'})
colors = {'Girls': "#eb8fbd", # pink
'Boys': "#3b8ee1"} # blue

fig = px.line(df_filter, y="OVERWEIGHT", x="YEAR", title="Overweight in Time", color="SEX", color_discrete_map=colors)
fig.update_yaxes(range=[0, 0.5])
fig.update_traces(fill="tozeroy")
fig.show()
```

```python
list_columns = ["SWEETS", "TOOTH_BRUSHING", "BREAKFAST_WEEKDAYS", "SOFT_DRINKS", "PHYS_ACT_60", "NERVOUS"]

df_2018 = df[df["YEAR"] == 2018]
df_2018 = df_2018.groupby("SEX", as_index=False)[list_columns].mean()
df_2018




```

```python

dict_scales = {
    "FRUITS": 7,
    "SOFT_DRINKS": 7,
    "SWEETS": 7,
    "VEGETABLES": 7,
    "FRIEND_TALK": 7,
    "TIME_EXE": 7,
    "PHYS_ACT_60": 7,
    "DRUNK_30": 7,
    "FAMILY_MEALS_TOGETHER": 6,
    "BREAKFAST_WEEKDAYS": 6,
    "TOOTH_BRUSHING": 5,
    "STUD_TOGETHER": 5,
    "BUL_OTHERS": 5,
    "BUL_BEEN": 5,
    "FIGHT_YEAR": 5,
    "INJURED_YEAR": 5,
    "HEADACHE": 5,
    "STOMACHACHE": 5,
    "FEEL_LOW": 5,
    "NERVOUS": 5,
    "SLEEP_DIF": 5,
    "DIZZY": 5,
    "THINK_BODY": 5,
    "TALK_MOTHER": 5,
    "TALK_FATHER": 5,
    "HEALTH": 4,
    "LIKE_SCHOOL": 4,
    "SCHOOL_PRESSURE": 4,
    "COMPUTER_NO": 4
}

for feature in list_columns:
   df_2018[feature] = df_2018[feature] / dict_scales[feature]

df_2018["SEX_STRING"] = df_2018["SEX"].map({1:"Boys", 2:"Girls"})

df_sex_long = df_2018.melt(
id_vars=["SEX_STRING"],
value_vars=list_columns,
var_name="FEATURE",
value_name="VALUE"
)

df_sex_long["Y"] = df_sex_long["FEATURE"] + " - " + df_sex_long["SEX_STRING"]


fig_horizontal_barchart = px.bar(
df_sex_long,
x="VALUE",
y="Y",
color="SEX_STRING",
# + jak sortovat - dopredu nebo muzem zde? - zkusit co poslal Vojta na discord 
orientation="h",
title="Features: Girls vs Boys, year = 2018"
)


fig_horizontal_barchart


```
