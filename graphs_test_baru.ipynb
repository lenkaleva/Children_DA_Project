```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('data.csv')
```

```python
df = data.copy()
df['SEX'] = df['SEX'].map({2: 'Girls', 1: 'Boys'})

agg = df.groupby(['YEAR','SEX'])['OVERWEIGHT'].mean().reset_index()
agg['OVERWEIGHT'] = agg['OVERWEIGHT']*100

colors = {'Girls': "#eb8fbd", # pink
'Boys': "#3b8ee1"} # blue

fig = px.line(agg, x='YEAR', y='OVERWEIGHT', color='SEX', color_discrete_map=colors, markers=True, title='Overweight % Girls vs Boys by Year')
fig.update_yaxes(range=[0, 100])
fig.update_traces(fill='tozeroy')
fig.show()
```
